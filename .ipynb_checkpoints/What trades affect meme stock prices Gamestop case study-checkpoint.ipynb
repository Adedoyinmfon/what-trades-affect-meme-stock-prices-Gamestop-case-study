```python
def standardise_date_time(dataframe):
    newGME_1 = dataframe.copy()
    
    newGME_1['Date-Time'] = newGME_1['Date-Time'].apply(lambda x: x[:29])
    newGME_1['Exchange_Time'] = newGME_1['Exchange_Time'][0][:18]
    
    newGME_1['Date_ExchangeTime'] = newGME_1.apply(lambda x: x['Date'] + 'T' + x['Exchange_Time'], axis=1)
    
    newGME_1['Date_ExchangeTime'] = pd.to_datetime(newGME_1['Date_ExchangeTime'])
    newGME_1['Date-Time'] = pd.to_datetime(newGME_1['Date-Time'])
    
    return newGME_1

def remove_data_noise(dataframe):
    '''
    Removing noisy data based on three rule:
    
    1. Transaction price is greater than the prevailing best ask price;
    2. The quoted bid price exceeds the quoted ask price;
    3. Observation occurs before 09:30:00hrs or after 16:00:00hrs Eastern Time 
    (this is to ensure that only normal trading day observations are included in your final dataset).
    
    # additional 
    4. Filter trade type that does not have bid price and ask price
    5. Filter transactions with no price and volume values 
    '''
    newGME_1 = dataframe.copy()
    
    # remove noise rule 1
    newGME_1 = newGME_1[~(newGME_1['Price'] > newGME_1['Ask Price'])]
    # remove noise rule 2
    newGME_1 = newGME_1[~((newGME_1['Type'] == 'Quote') & (newGME_1['Bid Price'] > newGME_1['Ask Price']))]
    # remove noise rule 3
    newGME_1 = newGME_1[(newGME_1.Exchange_Time >= '09:30:00') & (newGME_1.Exchange_Time <= '16:00:00')]
    # remove noise rule 4
    newGME_1 = newGME_1[~((newGME_1['Type'] == 'Trade') & (newGME_1['Bid Size'].isna()) & (newGME_1['Ask Price'].isna()))]
    newGME_1 = newGME_1[~(newGME_1['Price'].isna() & newGME_1['Volume'].isna())]
    
    return newGME_1.reset_index(drop=True)


newGME_2 = standardise_date_time(newGME)
newGME_2 = remove_data_noise(newGME_2)
newGME_2.shape
newGME_2.set_index('Date-Time', inplace=True)
newGME_2.head()


```

```python
newGME_2['Date'] = pd.to_datetime(newGME_2['Date'])
newGME_2['Exchange_Time'] = pd.to_datetime(newGME_2['Exchange_Time'])
newGME_2.info()
```

# Clean data, eliminate Nan values and calculate tick based on the price of data provided. The data is also grouped into Buyer initiated and seller initiated based on the tick values as indicated in code below

```python
# Set the initial value of 'Tick' to 0
newGME_2['Tick'] = 0

# Calculate the tick direction using vectorized operations
tick_direction = newGME_2['Price'].diff()

# Update the 'Tick' column based on the tick direction
newGME_2.loc[tick_direction > 0, 'Tick'] = 1
newGME_2.loc[tick_direction < 0, 'Tick'] = -1

# Forward-fill any remaining NaN values in the 'Tick' column
newGME_2['Tick'].fillna(method='ffill', inplace=True)

# Classify trades as buyer or seller initiated based on tick direction
newGME_2['Initiated'] = 'Neither'
newGME_2.loc[newGME_2['Tick'] >= 0, 'Initiated'] = 'Buyer Initiated'
newGME_2.loc[newGME_2['Tick'] < 0, 'Initiated'] = 'Seller Initiated'

# Drop the 'Tick' column
#newGME_2 = newGME_2.drop(['Tick'], axis=1)

# Separate the buyer-initiated and seller-initiated trades
buyer_initiated = newGME_2[newGME_2['Initiated'] == 'Buyer']['Price']
seller_initiated = newGME_2[newGME_2['Initiated'] == 'Seller']['Price']

#Print the first few rows of the updated dataframe
newGME_2.head()
```

# Calculate Mid-Quote and Effective Spread

```python
# Create a new column with the mid-quote of bid and ask price
nwGME["Mid_Quote"] = (nwGME["Bid Price"] + nwGME["Ask Price"]) / 2

# Convert the "Trade_Type" column to numeric values
nwGME["Type"] = pd.to_numeric(nwGME["Type"], errors="coerce")

# Calculate the effective spread
nwGME["Effective_Spread"] = 2 * nwGME["Type"] * (nwGME["Price"] - nwGME["Mid_Quote"]) / nwGME["Mid_Quote"].astype(float)

# Display the updated dataframe
print(nwGME.head())
```

# Mid-Quote Autocorrelations for intraday frequency(10s, 30s, 60s)

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.stattools import acf



# Calculate the returns
nwGME.loc[:, 'Returns'] = nwGME['Mid_Quote'].pct_change()
autocorr = acf(nwGME['Returns'].dropna(), nlags=150000)
# Group the data by stock-day
nwGME_grouped = nwGME.groupby(pd.Grouper(freq='D'))

# Loop through each stock-day and calculate the first-order return autocorrelations at various intraday frequencies
for name, group in nwGME_grouped:
    print('Stock-Day:', name)
    
    # Resample the data at various intraday frequencies
    for freq in ['10s', '30s', '60s']:
        print('Intraday Frequency:', freq)
        
        # Calculate the first-order return autocorrelation
        autocorr = acf(nwGME['Returns'].dropna(), nlags=150000)
        #autocorr = group['Returns'].resample(freq).apply(lambda x: x.autocorr())
        print('Autocorrelation:', autocorr)
```

The output displays autocorrelation values for stock prices on January 28, 2021, at different intraday frequencies (10s, 30s, 60s). Autocorrelation measures the correlation between stock prices at different time intervals within the same day. The provided values indicate the strength and direction of this correlation for various time lags.These results suggest that the stock's price movement has some degree of predictability, but the predictability decreases as the time interval between observations increases.

```python
import matplotlib.pyplot as plt

fig, axs = plt.subplots(3, figsize=(8, 8))

# Group the data by stock-day
nwGME_grouped = nwGME.groupby(pd.Grouper(freq='D'))

# Loop through each stock-day and calculate the first-order return autocorrelations at various intraday frequencies
for i, (name, group) in enumerate(nwGME_grouped):
    # Resample the data at various intraday frequencies
    for j, freq in enumerate(['10s', '30s', '60s']):
        # Calculate the first-order return autocorrelation
        autocorr = group['Returns'].resample(freq).apply(lambda x: x.autocorr())
        # Plot the autocorrelation
        axs[j].plot(autocorr, label=f'Stock-Day {i+1}')
        axs[j].set_title(f'Intraday Frequency: {freq}')
        axs[j].set_xlabel('Lag')
        axs[j].set_ylabel('Autocorrelation')
        axs[j].legend()

plt.tight_layout()
plt.show()
```

```python
pos_autocorr_count = len(autocorr[autocorr > 0])
neg_autocorr_count = len(autocorr[autocorr < 0])

print("Number of positive autocorrelations:", pos_autocorr_count)
print("Number of negative autocorrelations:", neg_autocorr_count)
```

Autocorrelation values range from -1 to 1, where -1 indicates a perfect negative correlation, 0 indicates no correlation, and 1 indicates perfect positive correlation. Negative values suggest that returns tend to move in the opposite direction of the previous time period. There is some degree of negative autocorrelation in the returns of the stock throughout the day at a 30-second intraday frequency. Autocorrelation measures the correlation of a variable with itself at different time lags to identify patterns or trends in the data. Autocorrelation values are similar at different intraday frequencies, indicating no significant variation in data patterns or trends over time intervals. The number of positive and negative autocorrelations is almost equal, indicating that the data may be relatively stable over time and less likely to experience significant changes in direction.

# Group trades by size, and make deductions on what trades move stock prices of meme stock based on the cumulative change deduced

```python
GME_1 = nwGME1_copy
# define a function to categorize the trades
def categorize_trades(volume):
#     if volume >= 10000:
#         return 'Large'
#     elif volume >= 499:
#         return 'Medium'
#     else:
#         return 'Small'
    
    if volume >= 10000:  # Considerably large volume 
        return 'Large'
    elif volume >= 500:  # Medium
        return 'Small'
    elif volume >= 100:   # Small
        return 'Medium'
    else:
        return 'nano'

# apply the function to the Volume column to create a new column for trade categories
GME_1['Trade Category'] = GME_1['Volume'].apply(categorize_trades)

# print the updated DataFrame
GME_1.head()
```

# Categorise the data and deduce values for Cumulative Price Change, Number of Trades, percentage of Trade, Volume, percentage of Volume

```python
#calculate cumulative price change in each trade category
price_change1 = GME_1.groupby('Trade Category')['Price Change'].sum()

# calculate number of trades in each trade category
n_trades = GME_1.groupby('Trade Category')['Price'].count()

# calculate percentage of trades in each trade category from total trades
pcent_trades = n_trades / len(GME_1) * 100

# calculate volume of each trade category
volume = GME_1.groupby('Trade Category')['Volume'].sum()

# calculate volume percentage of each trade category
pcent_volume = volume / GME_1['Volume'].sum() * 100

# create a summary table
categorised_data = pd.DataFrame({'Cumulative Price Change': price_change1,
                              'Number of Trades': n_trades,
                              '% of Trades': pcent_trades,
                              'Volume': volume,
                              '% of Volume': pcent_volume})


categorised_data.head()
```

# OLS regression

```python
import statsmodels.api as sm

import statsmodels.formula.api as smf
# y = mx + b   
# identify your dependent and independent variables
y = GME_1['Price']
X = GME_1[['Volume']]

# Add a constant term to the independent variables
X = sm.add_constant(X)

# Fit the multiple linear regression model
model = sm.OLS(y, X).fit()

# Print a summary of the regression results
print(model.summary())
```

This regression model is attempting to explain the variation in the dependent variable "Price" using the independent variable "Volume". The regression results indicate that there is no statistically significant relationship between the two variables, as the p-value for the "Volume" coefficient is 0.239, which is higher than the conventional threshold of 0.05.

The R-squared value of 0.000 suggests that the independent variable "Volume" does not explain any significant proportion of the variation in "Price".

The constant coefficient of 379.1807 represents the expected value of "Price" when "Volume" is zero. However, since the coefficient for "Volume" is not statistically significant, this value may not have any practical significance.

It is also worth noting that the condition number of 1.68e+03 is quite large, indicating that there may be issues of multicollinearity or numerical problems in the model. Overall, the results suggest that the variable "Volume" may not be a useful predictor of "Price" in this model.

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Load the data into a pandas DataFrame
data = GME_1.copy()

# Extract the variables
X = data['Volume']
y = data['Price']

# Reshape the variables to be 2-dimensional
X = X.values.reshape(-1, 1)
y = y.values.reshape(-1, 1)

# Create the model and fit it to the data
model = LinearRegression().fit(X, y)

# Print the coefficients of the regression line
print('Coefficients:', model.coef_)

# Print the R-squared score of the model
print('R-squared:', model.score(X, y))

# Plot the regression line
plt.scatter(X, y, color="blue")
plt.plot(X, model.predict(X), color="red", linewidth=2)
plt.title("Volume of Trades vs. Price of GME Stock")
plt.xlabel("Volume of Trades")
plt.ylabel("Price of Stock")
plt.show()
```
